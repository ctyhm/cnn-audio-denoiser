# 第十章 网络爬虫

```python
import requests
r = requests.get('http://www.dangdang.com/')
print(r.text)
```

```python
import re
pa = re.compile(r'<span>(.+?)</span>')
spans = pa.findall(r.text)
for text in spans:
    print(text)
```

```python
list_dangdang = spans[1:-14]
```

```python
r = requests.get('http://www.tianmao.com/')
print(r.text)
```

```python
pa = re.compile(r'<a href=.+?>(.+?)</a>')
spans2 = pa.findall(r.text)
for text in spans2:
    print(text)
```

```python
list_tmall = spans2[2:]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


label = ['dangdang', 'tmall']
color = ['gray', 'goldenrod']
size = [len(list_dangdang), len(list_tmall)]  
pie = plt.pie(size, colors=color, labels=label, shadow=True)
plt.axis('equal')
plt.legend() 
plt.show()
```

```python
import numpy as np

array_dangdang = np.array(list_dangdang)
array_tmall = np.array(list_tmall)
np.intersect1d(array_dangdang, array_tmall)
```

```python
np.setdiff1d(array_dangdang, array_tmall)
```

```python
np.setdiff1d(array_tmall, array_dangdang)
```

```python
key_words = ['手机', '图书', '数码', '汽车']
result = []
for key_word in key_words:
    temp_d = []
    temp_t = []
    for item in list_dangdang:
        result_d = re.search(key_word, item)
        if result_d != None:
            temp_d.append(item)
    for item in list_tmall:
        result_t = re.search(key_word, item)
        if result_t != None:
            temp_t.append(item)
    result.append(temp_d)
    result.append(temp_t)

result
```
